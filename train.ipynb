```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
pd.set_option('display.max_columns', None)
```

```python
keys = ["INCOME","TAX","DEPARTMENT","GOVT","INDIA","PERMANENT","ACCOUNT","NUMBER","SIGNATURE","GOVERNMENT","DOB",	"MALE",	"YEAR",	"BIRTH","UNIQUE","IDENTIFICATION","AUTHORITY","AADHAAR","EXPIRY","ISSUE","SEX","NATIONALITY","GIVEN","SURNAME","<<<", "PASSPORT","TYPE","COUNTRY","CODE"]
```

```python
for k in keys:
    print(k)
```

READ EACH FILE IN DIRECTORY and CREATE VECTOR. STORE AS CSV.

```python
directorylist = ["RawData/Passport","RawData/Aadhar","RawData/PAN"]
ls = []
for directory in directorylist:
    for filename in os.listdir(directory):
        with open(os.path.join(directory, filename)) as f:
            s = f.read().upper()
            d = {}
            for k in keys:
                if k in s:
                    d[k] = 1
                else:
                    d[k] = 0
            if "Passport" in directory:
                d['Y'] = 0
            elif 'Aadhar' in directory:
                d['Y'] = 1
            else:
                d['Y'] = 2

            ls.append(d)
pd.DataFrame(ls).to_csv("train.csv")
```

```python
print(ls)
```

### START TRAINING

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df = df.iloc[:,1:]
```

```python
df.describe().T
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df[df['Y']==2]
```

```python
X = df.drop(['Y'],axis=1)
y = df['Y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y ,
                                   random_state=104, 
                                   test_size=0.2, 
                                   shuffle=True)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
def prediction(X_test):
    return model.predict(X_test)
    
    
def cal_accuracy(y_test, y_pred):
      
    print("Confusion Matrix: ",
        confusion_matrix(y_test, y_pred))
      
    print ("Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
      
    print("Report : ",
    classification_report(y_test, y_pred))
```

```python
y_predict = prediction(X_test)
cal_accuracy(y_test,y_predict)
```

```python
y_predict = prediction(X_train)
cal_accuracy(y_train,y_predict)
```
